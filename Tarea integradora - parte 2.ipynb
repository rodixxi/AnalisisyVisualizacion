```python
%matplotlib inline
from scipy.stats import norm
import matplotlib.pyplot as plt
import pandas as pd

# Define parameters for normal distribution.
mu = 0
sigma = 5
rng = range(-20,30)
```

```python
def graphic_normal(mu, sigma, rng, n_examples=10000):
    # Generate normal distribution with given mean and standard deviation.
    dist = norm(mu, sigma)

    plt.figure(figsize=(10,8))
    plt.subplot(311) # Creates a 3 row, 1 column grid of plots, and renders the following chart in slot 1.
    plt.plot(rng, dist.pdf(rng), 'r', linewidth=2)
    plt.title('Probability density function of normal distribution')


    # Plot probability density function and of this distribution.
    plt.subplot(312)
    plt.plot(rng, dist.cdf(rng))
    plt.title('Cumulutative distribution function of normal distribution')

    # Draw 1000 samples from the random variable.
    sample = dist.rvs(size=n_examples)

    print("Sample descriptive statistics:")
    print(pd.DataFrame(sample).describe())

    # Plot a histogram of the samples.
    plt.subplot(313)
    plt.hist(sample, bins=100, normed=True)
    plt.plot(rng, dist.pdf(rng), 'r--', linewidth=2)
    plt.title('{} random samples from normal distribution'.format(n_examples))

    plt.tight_layout()  # Make space for titles 
```

```python
graphic_normal(mu, sigma, rng)
```



Se puede observar que es un gráfico con forma de campana acompañado de una simetría perceptible, lo cual confirma que es una distribución normal, partiendo de ello consideramos observar el comportamiento modificando los valores de la media (mu), desviación estándar (sigma) y el rango (rng). 

Se procedió a modificar los valores de forma independiente: 

Media
Se logró ver como se traslada la gráfica con forma de campana en el eje x sin presentar deformación, lo cual permite inferir que en las distribuciones normales la media presenta el máximo valor indicando que es el escenario con mayor posibilidad y el descenso de forma simétrica a los lados de la media, en la medida que se aleja de la media indica que los escenarios extremos son menos probables. 

Desviación estándar 
Al modificar los valores de sigma se logra ver como la anchura de la campana aumenta y disminuye en relación a la variación de dicho valor, esta variación indica la dispersión de los escenarios que se alejan o se acercan entre sí, aun cuando la media mantenga el mismo valor. Una forma de ver este comportamiento es asociarlo por ejemplo a la frecuencia con que pueden presentarse cada uno los escenarios con respecto a los adyacentes. 

Rango 
Alternando los valores del rango se puede captar que los extremos de la campana se acercan a cero en el eje Y sin lograr que alcance el valor 0, por otra parte el rango sí puede afectar la figura de la gráfica, permitiendo presentar un imagen parcial de la misma.
